```python
import pandas as pd

data = {'Name': ['Alice', 'BOB', 'Charlie', 'dAvE'], 'Age': [25, 30, 35, 40]}
df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)

df['Name'] = df['Name'].apply(lambda x: x.swapcase())

print("\nDataFrame after swapping cases in the 'Name' column:")
print(df)
```

```python
import matplotlib.pyplot as plt

file_path = 'D:/VS_code/Sample files/text.txt'  

x = []
y = []
with open(file_path, 'r') as file:
    for line in file:
        values = line.split()
        x.append(int(values[0]))
        y.append(int(values[1]))

plt.plot(x, y, color='blue', linewidth=2)

plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.title('Line from Text File Data')

plt.show()
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

ticker = 'GOOGL'
start_date = '2016-10-03'
end_date = '2016-10-07'

data = yf.download(ticker, start=start_date, end=end_date)

plt.figure(figsize=(10, 6))

plt.plot(data.index, data['Open'], label='Open', marker='o')
plt.plot(data.index, data['High'], label='High', marker='o')
plt.plot(data.index, data['Low'], label='Low', marker='o')
plt.plot(data.index, data['Close'], label='Close', marker='o')

plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.title('Alphabet Inc. Financial Data (October 3, 2016 - October 7, 2016)')
plt.legend()

plt.xticks(rotation=45)

plt.grid()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y1 = [2, 1, 4, 3, 5]
y2 = [1, 2, 4, 6, 8]
y3 = [5, 3, 9, 7, 13]

plt.plot(x, y1, label='Line 1', color='red', linewidth=2)
plt.plot(x, y2, label='Line 2', color='blue', linewidth=2)
plt.plot(x, y3, label='Line 3', color='green', linewidth=2)

plt.title('Multiple Lines with Legends')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.legend()

plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = x ** 2
y4 = np.sqrt(x)

plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 1)  
plt.plot(x, y1, color='blue', label='sin(x)')
plt.title('Sine Function')
plt.legend()
plt.grid()

plt.subplot(2, 2, 2)  
plt.plot(x, y2, color='red', label='cos(x)')
plt.title('Cosine Function')
plt.legend()
plt.grid()

plt.subplot(2, 2, 3)  
plt.plot(x, y3, color='green', label='x^2')
plt.title('Square Function')
plt.legend()
plt.grid()

plt.subplot(2, 2, 4)  
plt.plot(x, y4, color='purple', label='sqrt(x)')
plt.title('Square Root Function')
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

groups = ['A', 'B', 'C', 'D', 'E']
means_men = [22, 30, 35, 35, 26]
means_women = [25, 32, 30, 35, 29]

x = np.arange(len(groups))

bar_width = 0.35

plt.figure(figsize=(10, 6))
plt.bar(x - bar_width / 2, means_men, bar_width, label='Men', color='blue')
plt.bar(x + bar_width / 2, means_women, bar_width, label='Women', color='purple')

plt.xlabel('Group', fontsize=12)
plt.ylabel('Scores', fontsize=12)
plt.title('Scores by Group and Gender', fontsize=16)
plt.xticks(x, groups)
plt.legend()

plt.tight_layout()
plt.show()
```
