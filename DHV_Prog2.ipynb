```python
import matplotlib.pyplot as plt
import numpy as np

means_men = [22, 30, 35, 35, 26]
means_women = [25, 32, 30, 35, 29]
std_men = [4, 3, 4, 1, 5]
std_women = [3, 5, 2, 3, 3]

x = np.arange(len(means_men))

fig, ax = plt.subplots()
bars_men = ax.bar(x, means_men, yerr=std_men, capsize=5, label='Men', color='black')
bars_women = ax.bar(x, means_women, bottom=means_men, yerr=std_women, capsize=5, label='Women', color='purple')

ax.set_xlabel('Categories')
ax.set_ylabel('Values')
ax.set_title('Stacked Bar Plot with Error Bars')
ax.set_xticks(x)
ax.set_xticklabels(['Category 1', 'Category 2', 'Category 3', 'Category 4', 'Category 5'])
ax.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.rand(100)
y = np.random.rand(100)
sizes = np.random.rand(100) * 1000

plt.scatter(x, y, s=sizes, alpha=0.7, color='blue')
plt.title('Scatter Plot with Balls of Different Sizes')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```

```python
# 1. Scatter plot for three groups comparing weights and heights
import matplotlib.pyplot as plt
import numpy as np

group1_weights = np.random.randint(50, 70, 10)
group1_heights = np.random.randint(150, 170, 10)
group2_weights = np.random.randint(60, 80, 10)
group2_heights = np.random.randint(160, 180, 10)
group3_weights = np.random.randint(70, 90, 10)
group3_heights = np.random.randint(170, 190, 10)

plt.scatter(group1_weights, group1_heights, label='Group 1', color='blue')
plt.scatter(group2_weights, group2_heights, label='Group 2', color='green')
plt.scatter(group3_weights, group3_heights, label='Group 3', color='red')

plt.title('Scatter Plot for Weights and Heights')
plt.xlabel('Weight (kg)')
plt.ylabel('Height (cm)')
plt.legend()
plt.grid(True)
plt.show()
```
